```python
calpop = pd.read_csv('CAPOP.csv')
calpop['CAPOP'] = calpop['CAPOP'].map(lambda x: x * 1000)
calpop = calpop.loc[7:11,:]
calpop.sort_values('CAPOP', ascending = False, inplace = True)
num_calpop = calpop['CAPOP']
num_cal_1 = calpop['CAPOP'][10] - calpop['CAPOP'][9]
num_cal_2 = calpop['CAPOP'][9] - calpop['CAPOP'][8]
# print(num_cal_1, num_cal_2)
num_cal = (num_cal_1, num_cal_2)
mean_cal = np.mean(num_cal)
mean_cal += calpop['CAPOP'][10]
date_cal = ('2019-01-01')
df = pd.DataFrame({'DATE': date_cal, 'CAPOP': mean_cal}, index = [0])
df
```

```python
calpop.sort_values('CAPOP', ascending = True, inplace = True)
calpop = calpop.append(df, ignore_index = True, sort = True)

calpop
```

```python
date_r1 = pd.date_range(start='1/2/2016', end='12/31/2016')
mean_cal_1 = (calpop['CAPOP'][2] - calpop['CAPOP'][1])/365
cal_pop_1 = mean_cal_1 + calpop['CAPOP'][1]
print(mean_cal_1)
for i in date_r1:
    df_r1 = pd.DataFrame({'DATE': i, 'CAPOP': cal_pop_1}, index = [0])
    calpop = calpop.append(df_r1, ignore_index = True, sort = True)
#     print(calpop)
    cal_pop_1 += mean_cal_1
#     print(num_cal)
    
calpop
```
