```python
#redding-good 
dti = pd.date_range('2015-07-02', periods=1279, freq='d')

redding_request = requests.get('https://api.darksky.net/forecast/e9d8afefcdc0cda8d0a245218ec5db96/40.58759,-122.39261,2015-07-01T08:00:00?exclude=currently,hourly,flags')
# print(redding_request)
redding_request_j = redding_request.content.decode("utf-8")
# print(redding_request_j)
redding_request_d = json.loads(redding_request_j)
# print(redding_request_d)
dw = redding_request_d["daily"]['data']
df_redding = pd.DataFrame(dw)
# print(df_redding)

for d in dti:
    d = str(d)
    d = d.replace('00:00:00','').strip()
    redding_request = requests.get('https://api.darksky.net/forecast/e9d8afefcdc0cda8d0a245218ec5db96/40.58759,-122.39261,' + d + 'T08:00:00?exclude=currently,hourly,flags')
#     print(redding_request)
    redding_request_j = redding_request.content.decode("utf-8")
    redding_request_d = json.loads(redding_request_j)
    dw = redding_request_d["daily"]['data']
    df2 = pd.DataFrame(dw)
#     print(df2)
    df_redding = pd.concat([df_redding, df2], sort = True)

df_redding.reset_index(inplace = True, drop = True)
df_redding['time'] = df_redding['time'].map(lambda x: datetime.utcfromtimestamp(x).strftime('%m-%d-%Y'))
df_redding = df_redding.set_index('time')

df_redding.head()
```

```python
df_redding_trim = df_redding[['apparentTemperatureHigh',
                             'apparentTemperatureLow', 
                             'apparentTemperatureMax', 
                             'apparentTemperatureMin', 
                             'cloudCover',
                             'dewPoint', 
                             'humidity', 
                             'icon', 
                             'moonPhase', 
                             'precipIntensity',
                             'precipIntensityMax', 
                             'precipProbability', 
                             'pressure', 
                             'summary',
                             'sunriseTime', 
                             'sunsetTime', 
                             'temperatureHigh', 
                             'temperatureLow',
                             'temperatureMax',
                             'temperatureMin', 
                             'uvIndex',
                             'visibility', 
                             'windBearing',
                             'windGust',
                             'windSpeed']]
df_redding_trim.tail()
```
