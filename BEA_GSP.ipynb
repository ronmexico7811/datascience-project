```python
cal_gdp = pd.read_csv('download.csv')
cal_gdp_t = cal_gdp.T
cal_gdp_slim = cal_gdp_t.iloc[:, 0:1]
cal_gdp_slim.reset_index(inplace = True)
cal_gdp_slim.columns = cal_gdp_slim.iloc[0]
cal_gdp_slim.drop(cal_gdp_slim.index[0], inplace = True)
cal_gdp_slim
```

```python
gdp_cal_1 = cal_gdp_slim['All industry total'][35]-cal_gdp_slim['All industry total'][34]
gdp_cal_2 = cal_gdp_slim['All industry total'][34]-cal_gdp_slim['All industry total'][33]
gdp = (gdp_cal_1, gdp_cal_2)
gdp_ave = np.mean(gdp)
gdp_4 = cal_gdp_slim['All industry total'][35] + gdp_ave
date_cal = ('2018:Q4')
df = pd.DataFrame({'Description': date_cal, 'All industry total': gdp_4}, index = [0])
df
```

```python
cal_gdp_ = cal_gdp_slim.append(df, ignore_index = True, sort = True)
cal_gdp_
```

```python
cal_di = pd.Series(pd.date_range(start='1/1/2010', periods=36, freq='3M'))
cal_di
```

```python
cal_gdp_m = pd.concat([cal_gdp_, cal_di], axis = 1)
cal_gdp_m.rename(columns={0:'Date', 'All industry total':'GDP'},inplace=True)
# cal_gdp_m.set_index('Date', inplace = True)
cal_gdp_m.drop(['Description'], axis=1, inplace=True)
cal_gdp_m['GDP'] = cal_gdp_m['GDP'].astype(str).astype(float)
cal_gdp_m
```

```python
gdp_4 = cal_gdp_m['GDP'][35] + gdp_ave
date_cal = pd.to_datetime('2018-12-31')
df = pd.DataFrame({'GDP': gdp_4, 'Date': date_cal}, index = [0])
# print(df.info())
cal_gdp_m = cal_gdp_m.append(df, ignore_index = True, sort = True)
cal_gdp_m
```

```python
cal_gdp_m.set_index('Date', inplace = True)

gdp_daily = cal_gdp_m.resample('D').interpolate()
gdp_daily.info()
```

```python
cal_gdp_f = gdp_daily.iloc[1977:3257,]
cal_gdp_f.to_csv('bea_gdp.csv')
cal_gdp_f
```
